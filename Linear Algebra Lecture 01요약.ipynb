***1.2 ROW REDUCTION AND ECHELON FORMS***

* **ECHELON FORM (or row echelon form=REF)**
1. 모든 nonzero행은 all zero행 위에 있어야한다.
2. 각각의 leading entry는 대각선으로 존재해야한다.    
(cf.leading entry: nonzero인 행의 맨 왼쪽에 있는 요소를 말한다.)
3. leading entry 밑의 요소들은 다 0이어야 한다.

위 세 조건을 만족을 시키면 matrix는 echelon form이 된다. 여기서 밑의 두 가지 조건까지 만족 시키면, 그 matrix는 reduced row echelon form(RREF)이 된다.

(주의: 위의 세가지 조건을 다 만족시킨다는 가정하에)
4. leading entry가 모두 1의 값을 가진다.
5. 각각의 leading entry가 포함된 column의 나머지 요소들은 0의 값을 가진다.

ex) 
$\begin{bmatrix}
2 & 3 & 2 & 1 \\
0 & 1 & -4 & 8 \\
0 & 0 & 0 & 5/2
\end{bmatrix}$ and $\begin{bmatrix}
1 & 0 & 0 & 29 \\
0 & 1 & 0 & 16 \\
0 & 0 & 1 & 3
\end{bmatrix}$

위의 matrix에서 왼쪽 matrix는 REF이고, 오른쪽 matrix는 RREF라고 한다.
이때, REF는 row operation 과정에 따라 다양한 matrix가 존재하지만, RREF는 유일한 form을 가진다.

* Linear Combinations

주어진 벡터 $v_1, v_2, \cdots, v_p$, 주어진 스칼라 $c_1, c_2, \cdots, c_p$에 대하여 

$ y = c_1v_1+ \cdots + c_pv_p$ 꼴을 linear combination이라고 한다.

만약 $v_1, \cdots, v_p$이 $R^n$ 포함되었을때, $v_1, \cdots, v_p$의 모든 linear combination은 Span{$v_1, \cdots, v_p$라고 표기한다.

Span{$v_1, \cdots, v_p$}는 $v_1, \cdots, v_p$ 으로 만들수 있는 선형결합의 집합이라고 할 수 있다. 수학적으로는 $c_1v_1+ c_2v_2+ \cdots + c_pv_p(c_1, \cdots,c_p는 스칼라이다.)$

Span{**u**}는 기하학적으로 직선이 된다.

Span{**u, v**}은 기하학적으로 평면을 이루게 된다.(단, u벡터가 v벡터의 실수배가 아닐 경우)

***1.8 INTRODUCTION TO LINEAR TREANSFORMATIONS***

* Transformation은 n차원 벡터가 m차원 벡터로 맵핑을 시켜주는 과정이다. ($R^n$ -> $R^m$) 그래서 funtion 혹은 mapping이라고도 한다.

$T$ :$R^n$ -> $R^m$

* $R^n$에 존재하는 x를 $R^m$의 $T(x)$로 할당한다. 
* 이때, x가 존재하는 $R^n$를 정의역(domain)이라하고, $T(x)$가 속하는 $R^m$을 공역(codomain)이라고 한다. 그리고 mapping된 $T(x)$를 치역(range)라고 한다.

**Matrix Transformations**

linear transformation을 만들면 linear transform에 해당하는 matrix가 만들어지는데, 이 matrix를 standard matrix라고 한다. 이 standard matrix는 unique하다.

* Linear Transformations
1. $T($**u**+**v**$)=T($**u**$)+ T($**v**$)$
2. $T(c$**u**$)=c T($**u**$)$

 (all **u,v**가 T의 정의역 범위일때)

***1.9 THE MATRIX OF A LINEAR TRANSFORMATION***

## **THEOREM 10.**
$T$: $R^n$ -> $R^m$인 inear transformation.
라고 하면, $T($**x**$)=$ *A* **x**(for all **x** in $R^n$)을 만족하는 유일한 matrix *A*가 존재한다. 

1. onto
 
 :치역과 공역이 같아야한다. $R^n$에 있는 x가 $R^m$에 있는 모든 값을 가르켜야한다.
($R^m$=range) 
2. one-to-one

 :$R^n$에 있는 x가 $R^m$에 mapping될 때, 같은 $T(x)$를 가르키는 x가 없는 경우를 말한다. 즉 x가 다 다른 치역의 값을 가르킬 때, one-to-one이라고 한다.

## **THEOREM 11.**

$T$: $R^n$ -> $R^m$인 linear transformation
라고 하자. T가 one-to-one이다.  $\Leftrightarrow$ $T(x)=0$은 오직 trivial solution을 가진다.

$T(x)=0$은 linear하기 때문에 $T(0)=0$을 해로 가지는데, 겹치는 x값이 있으면 안되므로 **x**=**0**인 해만을 가진다. 따라서 trivial solution을 가진다.

## **THEOREM 12.**

$T$: $R^n$ -> $R^m$인 linear transformation
라고 하자.그리고 *A*는 이 transfomation의 standard matrix이다. 그러면 밑의 두 조건

1. *T*가 onto이다 $\Leftrightarrow$ $R^n$은 Span{*A* column}이다.
2. *T*가 one-to-one이면 $\Leftrightarrow$ *A*의 column들은 linearly independent하다.

 위의 Theorem 11에 의하면, T가 one-to-one일때 $T(x)=0$은 오직 trivial solution을 가지게 된다. trivial solution만을 가진다는 것은 linearly dependent하다라는 말과 같다.

즉, $T$가 one-to-one이다 $\Leftrightarrow$ $T(x)=0$은 오직 trivial solution을 가진다. $\Leftrightarrow$ *A* column들이 linearly dependent하다.

   

 cf) $T$가 one-to-one이면서 onto이면 즉, 치역과 공역이 같으면서 치역을 향하는 정의역이 겹치는게 없으면 standard matrix *A*는 역행렬(*A*$^-1$)을 가진다. 역행렬을 다루는 방법은 2장인 Matrix Algebra에서 다룬다.
